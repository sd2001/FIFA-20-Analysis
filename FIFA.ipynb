# GOALKEEPER



```python
a=2
b=1.5
c=1
d=0.5
```

```python
df['Goalkeeper']=(a*(df['GK Diving']+df['GK Kicking']+df['GK Positioning']+df['GK Reflexes'])+
                 b*(df['Long Passing']+df['Short Passing']+df['Reactions'])+
                 c*(df['Jumping']+df['Strength'])+
                 d*(df['Vision']))
print(df['Goalkeeper'])
```
